```python
file = './dbscripts/1_table.sql'  # Replace with your file name
file2 = './dbscripts/2_data.sql'
```

```python
with open(file, 'rb') as f:
    bom = f.read(3)
    if bom == b'\xef\xbb\xbf':  # UTF-8 BOM 的字节表示
        print("文件包含 UTF-8 BOM")
```

```python
with open(file, 'r', encoding='utf-8-sig') as f:
    content = f.read()
with open('converted_file.sql', 'w', encoding='utf-8') as f:
    f.write(content)  # 去除 BOM，保存为纯 UTF-8
```
